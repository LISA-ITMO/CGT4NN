# Logistic regression

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(133133)

Z = 5
W = 200

# Blob 1
X = np.random.standard_normal(W)
y = np.random.standard_normal(W)

# Blob 2
X2 = np.random.standard_normal(W) + Z
y2 = np.random.standard_normal(W) + Z

X3 = np.concatenate((X, X2), axis=0)
y3 = np.concatenate((y, y2), axis=0)

plt.scatter(X3, y3)
plt.show()
```

```python
b = 1
k = 1
j = 1

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def combine(x, y):
    return b + k * x + j * y

def predict(x, y):
    print(x, y)
    return sigmoid(combine(x, y))

i = 9
predict(X3[i], y3[i])
```
