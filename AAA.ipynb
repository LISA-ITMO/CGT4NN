```python
import matplotlib.pyplot as plt
import numpy as np

def plot_complex_numbers(complex_array, title="Complex Plane", x_label="Real", y_label="Imaginary"):
  """Plots an array of complex numbers on a complex plane using matplotlib.

  Args:
    complex_array: A NumPy array or list of complex numbers.
    title: (Optional) The title of the plot.
    x_label: (Optional) The label for the x-axis (real part).
    y_label: (Optional) The label for the y-axis (imaginary part).
  """

  # Extract real and imaginary parts
  real_parts = np.real(complex_array)
  imag_parts = np.imag(complex_array)

  # Create the plot
  plt.figure(figsize=(8, 8))  # Adjust figure size for better visualization
  plt.plot(real_parts, imag_parts, 'o')  # 'o' for circles as markers

  # Customize the plot
  plt.title(title)
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.grid(True)  # Add a grid for better readability

  # Ensure equal scaling of axes (important for the complex plane)
  plt.gca().set_aspect('equal', adjustable='box')

  # Add horizontal and vertical axes
  plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
  plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)

  # Automatically adjust axis limits
  # This is crucial for viewing all the points without manual adjustments
  # Find the maximum absolute value of real and imaginary parts
  max_abs = max(np.max(np.abs(real_parts)), np.max(np.abs(imag_parts)))
  padding = max_abs * 0.1  # Add a small padding for better visualization
  plt.xlim(-max_abs - padding, max_abs + padding)
  plt.ylim(-max_abs - padding, max_abs + padding)
  
  # Show the plot
  plt.show()

```

```python
plot_complex_numbers(y)
```
