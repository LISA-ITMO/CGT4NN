# Binarizer

https://sklearner.com/scikit-learn-binarizer/

```python
from sklearn.datasets import make_classification
from sklearn.preprocessing import Binarizer
import numpy as np

# generate continuous data
X, _ = make_classification(
    n_samples=100,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    random_state=1,
)
X = X * 10  # scale data to have wider range of values

# configure the binarizer
binarizer = Binarizer(threshold=0.0)

# transform the dataset
binary_X = binarizer.fit_transform(X)

# show before and after transformation
print("Before Binarization:")
print(X[:5])
print("After Binarization:")
print(binary_X[:5])
```

```python
import numpy as np

a = np.hstack((binary_X, X))
a
```
