# AugmentedReLUNetworkMultilayer

Можно указать любое количество слоёв

```python
from cgtnnlib.nn.AugmentedReLUNetworkMultilayer import AugmentedReLUNetworkMultilayer
from cgtnnlib.datasets import datasets

dataset = datasets['wine_quality']

model = AugmentedReLUNetworkMultilayer(
    inputs_count=dataset.features_count,
    outputs_count=dataset.classes_count,
    p=0.5,
    inner_layer_size=155,
    hidden_layers_count=99,
)
model
```

Also check out this compose function

```python
def compose(f, g):
    def wrapper(*args, **kwargs):
        return f(g(*args, **kwargs))
    return wrapper

def a(x):
    return x * 2

def b(x):
    return x / 3

compose(b, a)(4)
```
